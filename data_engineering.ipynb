```python
import pandas as pd

torvet= pd.read_csv("./data/torvet.csv",sep=";", header=None)
torvet=torvet.drop([5,3], axis=1)
for index in torvet.index:
    torvet.iloc[index,2]=torvet.iloc[index,2].replace('.','')
    if(torvet.iloc[index,2].isdigit()):
        torvet.iloc[index,2]=float(torvet.iloc[index,2])
    else:
        torvet.iloc[index,2]=None
        
    torvet.iloc[index,0]=torvet.iloc[index,0].replace('Audi','').strip()

bool_matrix=torvet.isnull() 
only_null_filter = bool_matrix.any(axis=1)
missing = torvet[only_null_filter]
torvet.dropna(inplace = True)
col=torvet.columns
torvet=torvet.rename(columns={col[0]: "model",col[1]: 'price',col[2]:"km",col[3]:'year'})
torvet.loc[1500:1600]
```
